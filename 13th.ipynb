```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')    #load sample dataset from internet
df.head()        
```

```python
df.columns=['total_bill','tip','sex','smoker','day','time','size']
df
```

relplot()
catplot()

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='sex',style='day',col='day')  #positive relationship data
#by default gives scatter plot
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='sex',style='day',col='day',kind='line') #change to line
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='violin')#for numerical or catbolical value
#import warnings
#warnings.filterwarnings('ignore')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
sns.barplot(x='day',y='total_bill',data=df,hue='sex')
```

```python
sns.catplot(x='day',y='total_bill',data=df,hue='sex',kind='bar')
plt.title('Khaoo peeo Dhaba')
plt.ylabel('total bill')
plt.xlabel('day')
plt.show()
```

```python
df['day'].value_counts()
key=list(df['day'].value_counts().keys())
value=list(df['day'].value_counts().values)
plt.bar(key,value)
```

```python
plt.pie(value,labels=key)
plt.show()
```

```python
plt.pie(value,labels=key,wedgeprops={'ec':'k'})
plt.show()
```

```python
ls=[0.5,0.4,0.3,0.2]
plt.pie(value,labels=key,wedgeprops={'ec':'k'},explode=ls,autopct="%1.2f%%",shadow=True)
plt.title('khao peeo dhaba',loc='left')
sns.set_theme(rc={'figure.figsize':(8,8)})
plt.savefig("khaopeeodhaba")
plt.show()
```

```python
sns.catplot(x='time',y='total_bill',data=df,hue='smoker',kind='bar')
plt.title('khao peeo dhaba',loc='center')
plt.show()
```

```python
sns.scatterplot(x='size',y='total_bill',data=df)
```
