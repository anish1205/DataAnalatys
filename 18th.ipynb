```python
df=pd.read_csv('cleaned_data.csv')
df.head()
```

```python
x=df.drop('Customer Type_disloyal Customer',axis='columns')
y=df[['Customer Type_disloyal Customer']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split    
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

logistic regression algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
x_train
```

```python
# import warnings
# warnings.filterwarnings('ignore') 
```

```python
lg.fit(x_train,y_train)

```

```python
lg.score(x_train,y_train)
```

```python
lg.fit(x_test,y_test)
```

```python
lg.score(x_test,y_test)
```

```python
round(lg.score(x_train,y_train),2)*100
```

```python
round(lg.score(x_test,y_test),2)*100
```

```python
pred=lg.predict(x_test)
```

```python
y_test['prediction']=pred
```

```python
from sklearn.metrics import confusion_matrix 
```

```python
confusion_matrix(y_test['Customer Type_disloyal Customer'],y_test['prediction'])
```

         +ve     -ve
        -------------
 +ve    | TP  |  FP  |      TP ---> 1    1
        --------------
 -ve    | FN  | TN   |      TN----> 1    0
        --------------      FP----> 0    0
        confusion matrix    FN----> 0    1


```python
A,B = confusion_matrix(y_test['Customer Type_disloyal Customer'],y_test['prediction'])
```

```python
TP= A[0]
FP=A[1]
FN=B[0]
TN=B[1]
```

```python
accuracy=(TP+TN)/(TP+TN+FP+FN)
```

```python
accuracy
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test['Customer Type_disloyal Customer'],y_test['prediction']))
```
